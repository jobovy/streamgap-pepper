```python
import os.path
def compute_actions(data_rand,centre,fname):
    if os.path.isfile(fname):
        return pd.read_csv(fname)
    aa = aa_py.Actions_Genfunc(pot,"axisymmetric")
#     aa = aa_py.Actions_AxisymmetricStackel_Fudge(pot,-30.)
    data_rand_aa = pd.DataFrame(data_rand)
    data_rand_aa['JR'] = np.nan
    data_rand_aa['Lz'] = np.nan
    data_rand_aa['Jz'] = np.nan
    data_rand_aa['TR'] = np.nan
    data_rand_aa['Tp'] = np.nan
    data_rand_aa['Tz'] = np.nan
    data_rand_aa['OR'] = np.nan
    data_rand_aa['Op'] = np.nan
    data_rand_aa['Oz'] = np.nan
    for n,i in enumerate(data_rand.values):
        acts = aa.actions(i[:6])
        angs = aa.angles(i[:6])
        data_rand_aa['JR'][n]=acts[0]
        data_rand_aa['Lz'][n]=acts[1]
        data_rand_aa['Jz'][n]=acts[2]
        data_rand_aa['TR'][n]=angs[0]
        data_rand_aa['Tp'][n]=angs[1]
        data_rand_aa['Tz'][n]=angs[2]
        data_rand_aa['OR'][n]=angs[3]
        data_rand_aa['Op'][n]=angs[4]
        data_rand_aa['Oz'][n]=angs[5]
    centre_aa = aa.angles(centre)
    centre_aa2 = aa.actions(centre)
    data_rand_aa['dangR']=data_rand_aa['TR']-centre_aa[0]
    data_rand_aa['dangp']=data_rand_aa['Tp']-centre_aa[1]
    data_rand_aa['dangz']=data_rand_aa['Tz']-centre_aa[2]
    data_rand_aa['dangR']-=2.*np.pi*(data_rand_aa['dangR']>np.pi)
    data_rand_aa['dangp']-=2.*np.pi*(data_rand_aa['dangp']>np.pi)
    data_rand_aa['dangz']-=2.*np.pi*(data_rand_aa['dangz']>np.pi)
    data_rand_aa['dOmR']=data_rand_aa['OR']-centre_aa[3]
    data_rand_aa['dOmp']=data_rand_aa['Op']-centre_aa[4]
    data_rand_aa['dOmz']=data_rand_aa['Oz']-centre_aa[5]
    data_rand_aa['dJR']=data_rand_aa['JR']-centre_aa2[0]
    data_rand_aa['dLz']=data_rand_aa['Lz']-np.fabs(centre_aa2[1])
    data_rand_aa['dJz']=data_rand_aa['Jz']-centre_aa2[2]
    
    n = sampler.par_vector()
    
    data_rand_aa['dOmpar']=np.sum(np.vstack((data_rand_aa['dOmR'],data_rand_aa['dOmp'],data_rand_aa['dOmz'])).T*n,axis=1)
    data_rand_aa['dOmperp']=np.sqrt(np.sum(np.vstack((data_rand_aa['dOmR'],data_rand_aa['dOmp'],data_rand_aa['dOmz'])).T**2,axis=1)-data_rand_aa['dOmpar']**2)
    data_rand_aa['dangpar']=np.sum(np.vstack((data_rand_aa['dangR'],data_rand_aa['dangp'],data_rand_aa['dangz'])).T*n,axis=1)
    data_rand_aa['dangperp']=np.sqrt(np.sum(np.vstack((data_rand_aa['dangR'],data_rand_aa['dangp'],data_rand_aa['dangz'])).T**2,axis=1)-data_rand_aa['dangpar']**2)
    data_rand_aa.to_csv(fname)
    return data_rand_aa
```

```python
def make_sample(sampler,N = 6400,lt=0,dT=np.array([]),T=np.array([])):
    results = np.zeros(shape=(N,25))
    if(len(T)==0):
        st = "linear"
        cumul=np.array([])
    else:
        st = "cumul"
        cumul = sampler.form_cumulative(T,dT)
    for i in range(len(results)):
        results[i]=sampler.sample(lt,st,cumul,dT)
    return pd.DataFrame(results,
                           columns=['x','y','z','vx','vy','vz',
                                    'OmR','Omp','Omz','ThR','Thp','Thz',
                                    'DOmR','DOmp','DOmz','DThR','DThp','DThz',
                                    'Ompar','Omperp','Thpar','Thperp',
                                    'JR','Lz','Jz'])
def sample_eq(sampler,N=6400,lt=0,dT=np.array([]),T=np.array([]),use_Denis=False):
    results = make_sample(sampler,N,lt,dT,T)
    SM = SolarMotion
    if(use_Denis):
        SM=SolarMotion_D
    eq = np.array(map(lambda x,y,z,vx,vy,vz:
                aa_py.GalacticToEquatorial(
                aa_py.CartesianToGalactic(
                    np.array([x,y,z,vx,vy,vz]),
                    SM)),
                results.x,results.y,results.z,
                    results.vx,results.vy,results.vz))
    results['ra']=eq.T[0]
    results['dec']=eq.T[1]
    results['dist']=eq.T[2]
    results['vlos']=eq.T[3]
    results['pmra']=eq.T[4]
    results['pmdec']=eq.T[5]
    return results
def compute_eq(results,use_Denis=False):
    SM = SolarMotion
    if(use_Denis):
        SM=SolarMotion_D
    return np.array(map(lambda x,y,z,vx,vy,vz:
                aa_py.GalacticToEquatorial(
                aa_py.CartesianToGalactic(
                    np.array([x,y,z,vx,vy,vz]),
                    SM)),
                results.x,results.y,results.z,
                    results.vx,results.vy,results.vz))
```

```python
ra = 229.0208 ## deg
dec = -0.1117## deg
vlos = -58.7 ## km/s
dist = 23. ## kpc
pmra = -2.296 ## mas/yr
pmde = -2.257 ## mas/yr
# Coordinate transformation routines
_RAPAL5= 229.018/180.*numpy.pi
_DECPAL5= -0.124/180.*numpy.pi
_TPAL5= numpy.dot(numpy.array([[numpy.cos(_DECPAL5),0.,numpy.sin(_DECPAL5)],
                               [0.,1.,0.],
                               [-numpy.sin(_DECPAL5),0.,numpy.cos(_DECPAL5)]]),
                  numpy.array([[numpy.cos(_RAPAL5),numpy.sin(_RAPAL5),0.],
                               [-numpy.sin(_RAPAL5),numpy.cos(_RAPAL5),0.],
                               [0.,0.,1.]]))

def radec_to_pal5xieta(ra,dec,degree=False):
    XYZ= numpy.array([numpy.cos(dec)*numpy.cos(ra),
                      numpy.cos(dec)*numpy.sin(ra),
                      numpy.sin(dec)])
    phiXYZ= numpy.dot(_TPAL5,XYZ)
    phi2= numpy.arcsin(phiXYZ[2])
    phi1= numpy.arctan2(phiXYZ[1],phiXYZ[0])
    return numpy.array([phi1,phi2]).T
```

### Computation of power spectrum and wavelet scalogram

```python
def fit_polynomial(y,x,n):
    coeffs = np.ones(n)
    coeffs[-1]=np.mean(n)
    def residual(p,y,x):
        return (y-np.polyval(p,x))**2/y
    return leastsq(residual,coeffs,args=(y,x))[0]
```

```python
from scipy.signal import csd
def compute_jo_power_spectrum(data,binsize=0.1,minn=0.,maxx=18.,data2=None,noise=False,fitoutpolynomial=-1):
    bins=np.arange(minn,maxx,binsize)
    bins_c = .5*(bins[1:]+bins[:-1])
    n,b=np.histogram(data,bins=bins)
    if(fitoutpolynomial>0):
        pol = fit_polynomial(n,bins_c,fitoutpolynomial)
        n = n/np.polyval(pol,bins_c)
    if(data2!=None):
        n2,b2=np.histogram(data2,bins=bins)
        if(fitoutpolynomial>0):
            pol = fit_polynomial(n2,bins_c,fitoutpolynomial)
            n2 = n2/np.polyval(pol,bins_c)
        n = n.astype(float)/n2.astype(float)
    px = csd(n,n,fs = 1./(b[1]-b[0]),scaling='spectrum',nperseg=len(n))
    oversamples=100
    if(noise):
        pnoises=np.empty((oversamples,len(px[0])))
        for i in range(oversamples):
            noise = np.random.uniform(low=minn,high=maxx,size=len(data))
            nnoise = np.histogram(noise,bins=bins)[0]
            if(fitoutpolynomial>0):
                pol = fit_polynomial(nnoise,bins_c,fitoutpolynomial)
                nnoise=nnoise/np.polyval(pol,bins_c)
            elif(data2!=None):
                nnoise=nnoise*(1./len(oversamples*data)*(maxx-minn)/binsize)
            pnoises[i] = csd(nnoise,nnoise,fs = 1./(b[1]-b[0]),scaling='spectrum',nperseg=len(n))[1]
        pnoise=np.nanmedian(pnoises,axis=0)
        pnoise_s=np.nanpercentile(pnoises,15.,axis=0)
        pnoise_s2=np.nanpercentile(pnoises,85.,axis=0)
        return 1./px[0],np.sqrt(px[1].real*(b[-1]-b[0])),np.sqrt(pnoise.real*(b[-1]-b[0])),np.sqrt(pnoise_s.real*(b[-1]-b[0])),np.sqrt(pnoise_s2.real*(b[-1]-b[0]))
    else:
        return 1./px[0],np.sqrt(px[1].real*(b[-1]-b[0]))
```

```python
from astroML.fourier import wavelet_PSD

def compute_wavelet_power_spectrum(data,binsize=0.1,minn=0.,maxx=18.,data2=None,
                                   noise=False,log_sp=False,fitoutpolynomial=-1,frange=[0.1,10.],
                                   Q=0.25,test=False):
    
    bins=np.arange(minn,maxx+binsize,binsize)
    bins_c = .5*(bins[1:]+bins[:-1])
    n,b=np.histogram(data,bins=bins)
    
    if(test):
        noiser = np.random.uniform(low=minn,high=maxx,size=len(data[(data>minn)&(data<maxx)]))
        nnoise = np.histogram(noiser,bins=bins)[0]
        n = 30.*np.sin(2.*np.pi*bins_c)*np.exp(-(bins_c-5.)**2/2./Q**2)+nnoise
        
    polyvals=np.zeros(len(bins_c))
    if(fitoutpolynomial>0):
        pol = fit_polynomial(n,bins_c,fitoutpolynomial)
        polyvals = np.polyval(pol,bins_c)
        n = n/polyvals-1.
    
    f0 = np.linspace(frange[0],frange[1], 100)
    if(log_sp):
        f0 = np.logspace(np.log10(frange[0]),np.log10(frange[1]), 100)
    db = .5*(bins_c[-1]-bins_c[0])
    bins_c-=db
    wPSD = wavelet_PSD(bins_c,n, f0, Q=Q)
    
    oversamples=100
    if(noise):
        pnoises=np.empty((oversamples,len(f0)))
        for i in range(oversamples):
            noise = np.random.uniform(low=minn,high=maxx,size=len(data[(data>minn)&(data<maxx)]))
            nnoise = np.histogram(noise,bins=bins)[0]
            nnoise = nnoise.astype(float64)
            if(fitoutpolynomial>0):
                pol = fit_polynomial(nnoise,bins_c,fitoutpolynomial)
                nnoise=nnoise/np.polyval(pol,bins_c)-1.
            elif(data2!=None):
                nnoise=nnoise*(1./len(oversamples*data)*(maxx-minn)/binsize)
            pnoises[i] = np.sum(wavelet_PSD(bins_c,nnoise, f0, Q=Q),axis=1)*(bins_c[1]-bins_c[0])
        pnoise=np.nanmedian(pnoises,axis=0)
        pnoise_s=np.nanpercentile(pnoises,15.,axis=0)
        pnoise_s2=np.nanpercentile(pnoises,85.,axis=0)
        return bins_c+db,f0,n,wPSD,pnoise,pnoise_s,pnoise_s2,polyvals
    return bins_c+db,f0,n,wPSD

def make_plot_wpsd(asda,log_t=False,log_sp=False,vrange=[None,None]):
    plt.figure(figsize=[16.,4.])
    plt.subplot(1,4,1)
    if(log_t):
        plt.imshow(np.log(asda[3]),origin='lower',aspect='auto',cmap=plt.cm.viridis,extent=[asda[0][0],asda[0][-1],asda[1][0],asda[1][-1]],
                   vmin=vrange[0],vmax=vrange[1])
    else:
        plt.imshow(asda[3],origin='lower',aspect='auto',cmap=plt.cm.viridis,extent=[asda[0][0],asda[0][-1],asda[1][0],asda[1][-1]],
                   vmin=vrange[0],vmax=vrange[1])
    if(log_sp):
        plt.semilogy()
    plt.axvline(5.)
#     plt.contour(np.log10(asda[3]),origin='lower',aspect='auto',cmap=plt.cm.viridis,extent=[asda[0][0],asda[0][-1],asda[1][0],asda[1][-1]])
    plt.xlabel(r'$\xi$')
    plt.ylabel(r'$f_\xi$')
    plt.xlim(plt.gca().get_xlim()[::-1])
    plt.subplot(1,4,2)
    plt.plot(asda[1],np.sum(asda[3],axis=1)*(asda[0][1]-asda[0][0]))
    plt.plot(asda[1],asda[4],color='k')
    plt.fill_between(asda[1],asda[5],asda[6],color='gray',alpha=0.5)
    plt.semilogy()
    plt.xlabel(r'$f_\xi$')
    plt.subplot(1,4,3)
    plt.plot(asda[1],np.sum(asda[3],axis=1)*(asda[0][1]-asda[0][0])/asda[4])
    plt.xlabel(r'$f_\xi/f_\mathrm{noise}$')
    plt.subplot(1,4,4)
    plt.plot(asda[0],asda[2])
    if(log_sp):
        plt.plot(asda[0],np.sum((asda[3].T*log(10.)*asda[1]).T,axis=0)*(np.log10(asda[1][1])-np.log10(asda[1][0])))
    else:
        plt.plot(asda[0],np.sum(asda[3],axis=0)*(asda[1][1]-asda[1][0]))
    plt.xlabel(r'$\xi$')
    plt.xlim(plt.gca().get_xlim()[::-1])
```

```python
orb = aa_py.Orbit(pot,1e-10)
def integrate(x,t,frac=1,dt=None):
    if(dt==None):
        dT = frac*pot.torb(x)
    else:
        dT=dt
    r = np.zeros((t/dT+2,6))
    r[0]=x
    i,T=1,0
    while(T<t):
        r[i]=orb.integrate(r[i-1],dT,dT,False)
        T+=dT
        i+=1
    return r[:-1],T
```
